```python
sks['REVENUE'] = sks['RETAIL']-sks['COST']
df_trnsact_sks = df_trnsact.merge(sks,on = ['SKU','STORE'], how = 'left')
df_trnsact_sks
```

```python
df_trnsact_sks['DISCOUNT'] = df_trnsact_sks['RETAIL'] - df_trnsact_sks['ORIGPRICE']
df_trnsact_sks
```

```python
def create_time_features(df):
    df = df.copy()
    df['quarter'] = pd.PeriodIndex(df['SALEDATE'], freq='Q')
    df['month'] = pd.PeriodIndex(df['SALEDATE'], freq='M')
    df['year'] = pd.PeriodIndex(df['SALEDATE'], freq='Y')
    df['weekofyear'] = pd.PeriodIndex(df['SALEDATE'], freq='D')
    return df
```

```python
df_trnsact_sks.SALEDATE = pd.to_datetime(df_trnsact_sks.SALEDATE)
df = create_time_features(df_trnsact_sks)
df = pd.get_dummies(df, prefix=['STYPE'], columns=['STYPE'])
df_groupyed = df.groupby(['STORE', 'month']).agg({'SKU': 'count', 'STYPE_P': 'sum','STYPE_R':'sum','AMT':'sum','REVENUE':'sum','DISCOUNT':'sum' })
df_groupyed = df_groupyed.reset_index(level = [0,1])
df_groupyed
```

```python
df_new = pd.DataFrame(columns = df_groupyed.columns)
def create_lag_feature(df):
    df['AMT_1'] = df['AMT'].shift(1)
    df['AMT_3'] = df['AMT'].shift(3)
    df['AMT_6'] = df['AMT'].shift(6)
    df['REVENUE_1'] = df['REVENUE'].shift(1)
    df['REVENUE_3'] = df['REVENUE'].shift(3)
    df['REVENUE_6'] = df['REVENUE'].shift(6)
    df['DISCOUNT_1'] = df['DISCOUNT'].shift(1)
    df['DISCOUNT_3'] = df['DISCOUNT'].shift(3)
    df['DISCOUNT_6'] = df['DISCOUNT'].shift(6)
    
    return df
```

```python
for store in df_groupyed['STORE'].drop_duplicates().tolist():
    df_part = df_groupyed[df_groupyed['STORE']==store]
    df_new = df_new.append(create_lag_feature(df_part))
df_new
```
